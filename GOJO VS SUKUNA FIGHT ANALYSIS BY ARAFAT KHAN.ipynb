```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\HP\Downloads\jjk_fight_analysis.csv")
```

```python
Avg_damage = df.groupby("character") ["damage_level"].mean()
Avg_damage
```

```python
Avg_energy = df.groupby("character") ["energy_cost"].mean()
Avg_energy
```

```python
Adapt_moves = df[df["adaptation_required"] == "Yes"]
Adapt_moves
```

```python
Domain_usage = df[df["domain_used"] == "Yes"].groupby("character").size()
Domain_usage
```

```python
Fight_phase_analysis = df.groupby("fight_phase") ["damage_level"].mean()
Fight_phase_analysis
```

```python
import matplotlib.pyplot as plt 
```

```python
Avg_damage = df.groupby("character") ["damage_level"].mean()
Avg_damage.plot(kind = "bar",
                 title = "Average Damage by Character",
                 ylabel = "Average level of damage",
                 xlabel = "Character")
plt.show()
```

```python
Avg_energy_cost = df.groupby("character") ["energy_cost"].mean()
Avg_energy_cost.plot(kind = "bar",
                 title = "Average Energy Cost by Character",
                 ylabel = "Average level of energy cost",
                 xlabel = "Character")
plt.show()
```

```python
Adapt_count = Adapt_moves["character"].value_counts()
Adapt_count.plot(kind = "bar",
                 title = "Adaptive Moves Damage Analysis",
                 ylabel = "Number of Moves",
                 xlabel = "Character")
plt.show()
```

```python
df["efficiency"] = df["damage_level"] / df["energy_cost"]

efficiency = df.groupby("character")["efficiency"].mean()

efficiency.plot(kind="bar",
                 ylabel="Average Damage per Energy",
                 xlabel="Character",
                 title="Damage per Energy (Efficiency)")
plt.show()
```
